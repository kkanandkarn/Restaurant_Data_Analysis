# I Want to Know who gave the highest tip ?

```python
# Highest tip
Female_htip = sum(restra1.tip[restra1.sex==0])
Male_htip = sum(restra1.tip[restra1.sex==1])

print("Highest tip given by Women is: ", Female_htip)
print("Highest tip given by Men is: ", Male_htip)

# Maximum tip
Female_mtip = max(restra1.tip[restra1.sex==0])
Male_mtip = max(restra1.tip[restra1.sex==1])

print("\nMaximum tip given by Women is: ", Female_mtip)
print("Maximum tip given by Men is: ", Male_mtip)
```

```python
# Calculate the highest tip and maximum tip given by women and men
Female_htip = sum(restra1.tip[restra1.sex == 0])
Male_htip = sum(restra1.tip[restra1.sex == 1])
Female_mtip = max(restra1.tip[restra1.sex == 0])
Male_mtip = max(restra1.tip[restra1.sex == 1])

# Set the data and labels
categories = ['Female', 'Male']
highest_tips = [Female_htip, Male_htip]
maximum_tips = [Female_mtip, Male_mtip]

# Set the positions of the bars
bar_width = 0.35
r1 = np.arange(len(categories))

# Create a grouped horizontal bar chart
plt.barh(r1, highest_tips, height=bar_width, label='Highest Tip', color='red')
plt.barh(r1, maximum_tips, height=bar_width, label='Maximum Tip', color='blue', alpha=0.7)

# Add labels, title, and legend
plt.xlabel('Amount')
plt.ylabel('Gender')
plt.title('Tips by Gender')
plt.yticks(r1, categories)
plt.legend()

# Add values on the bars
for i, v in enumerate(highest_tips):
    plt.text(v, i, str(v), color='black', va='center')
for i, v in enumerate(maximum_tips):
    plt.text(v, i, str(v), color='black', va='center')

# Display the chart
plt.show()
```

# Total bill amount per day.

```python
df = pd.read_csv('tips.csv')
```

```python
total_bill_earned = sum(restra1.total_bill)
print("Total bill eared: ", total_bill_earned)
```

# Sum of total_bills per day

```python
df.groupby(by=['day']).total_bill.agg(['sum'])
```

```python
# Calculate the sum of total bills grouped by day
df_grouped = df.groupby(by=['day']).total_bill.agg(['sum'])


# Create a pie chart
labels = df_grouped.index
values = df_grouped['sum']

plt.pie(values, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title('Total Bills by Day')

# Display the chart
plt.show()
```
